```python
apprMax = df['MaxTemp'].mean()
apprMin = df['MinTemp'].mean()
apprMax, apprMin
```

```python
locations = []
for location in df['Location'].unique():
    tempMaxAppr = df[df.Location == location]['MaxTemp'].mean()
    tempMinAppr = df[df.Location == location]['MinTemp'].mean()
    SKO_Max = df[df.Location == location]['MaxTemp'].std()
    SKO_Min = df[df.Location == location]['MinTemp'].std()
    if (tempMaxAppr > apprMax + SKO_Max) | (tempMinAppr < apprMin - SKO_Min):
        locations.append(location)
#print(locations)
dfOnlyUnOrderly = df[df.Location.isin(locations)]
dfOnlyUnOrderly
```

```python
#Отличающиеся от среднего на велиичину СКО
dfOnlyUnOrderly.boxplot(column = 'MaxTemp', by = 'Location', figsize = (10, 7));
```

```python
#Z-нормализация по MaxTemp
MO = df['MaxTemp'].mean()
SKO = df['MaxTemp'].std()
df_norm = df.copy()
df_norm = df_norm.dropna(how = 'any' , axis = 0)
df_norm['MaxTemp'] = (df_norm['MaxTemp']-MO)/SKO
df_norm
```

```python
#Z-нормализация по MinTemp
MO = df['MinTemp'].mean()
SKO = df['MinTemp'].std()
df_norm['MinTemp'] = (df_norm['MinTemp']-MO)/SKO
df_norm
```
