```python
def MultiColumnLabelEncoder(columns,df):
    X = df.iloc[:].values
    for i in range(len(columns)):
        X[:, columns[i]] = labelencoder.fit_transform(X[:, columns[i]])
    df = pd.DataFrame(X, columns=df.columns, index = df.index)
    return df
```

```python
import datetime
days = []
months = []
years = []
for date_string in df.Date:
    date_object = datetime.datetime.strptime(date_string, '%Y-%m-%d')
 
    days.append(date_object.day)
    months.append(date_object.month)
    years.append(date_object.year)
df['Month'] = months

cols = df.columns.tolist()
cols = cols[-3:-1] + [cols[-1]] + cols[:-3]

df=df[cols]
df
```

```python
df.drop(columns=['Date'], inplace=True)
df.info()
```

```python
df = df.dropna(how = 'any' , axis = 0)
```

```python
columns = [0, 3, 9, 11, 12]
df = MultiColumnLabelEncoder(columns,df)
df
```

```python
#Z-нормализация
for col in df.columns[2::]:
    MO = df[col].mean()
    SKO = df[col].std()
    df[col] = (df[col]-MO)/SKO
df
```

```python
X = df[['Humidity3pm','Sunshine']]
y = df['MaxTemp']
```

```python
#Разделение на тестовые и тренировочные данные
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
#Линейная регрессия
from sklearn.linear_model import LinearRegression
linRegModel = LinearRegression()
linRegModel.fit(X_train, y_train)
```

```python
y_pred = linRegModel.predict(X_test)
```

```python
from sklearn import metrics
print('R^2: ', metrics.r2_score(y_test, y_pred))
print('mean_squared_error: ', metrics.mean_squared_error(y_test, y_pred)) 
print('mean_absolute_error: ', metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 4))

ax.scatter(X_test['Humidity3pm'], y_test, color='grey')
ax.scatter(X_test['Humidity3pm'], y_pred, color='black')
```

```python
plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 4))

ax.scatter(X_test['Sunshine'], y_test, color='grey')
ax.scatter(X_test['Sunshine'], y_pred, color='black')
```

```python
#LASSO
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
parametrs = {'alpha': [0.1, 0.5, 1, 10], 'max_iter': [100, 500, 1000]}
LassoModel = Lasso()
grid_search = GridSearchCV(LassoModel, parametrs, cv=5)
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
```

```python
LassoModel = Lasso(alpha=0.1,max_iter=100)
LassoModel.fit(X_train, y_train)
y_pred = LassoModel.predict(X_test)
```

```python
print('R^2: ', metrics.r2_score(y_test, y_pred))
print('mean_squared_error: ', metrics.mean_squared_error(y_test, y_pred)) 
print('mean_absolute_error: ', metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
#Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
parametrs = {'alpha': [0.1, 0.5, 1, 10, 100], 'max_iter': [100, 500, 1000]}
RidgeModel = Ridge()
grid_search = GridSearchCV(RidgeModel, parametrs, cv=5)
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
```
